```python
import os
import json
import pandas as pd
from pprint import pprint
from copy import deepcopy
```

```python
final_df = pd.DataFrame()
data = {}
```

```python
final_df
```

```python
for file in os.listdir('../data/webmineral'):
    if file.startswith('FullMineralData'):
        with open(f'../data/webmineral/{file}', 'r') as f:
            data = json.load(f)

        new_data = []

        for k, v in data.items():
            new_dic = v
            new_dic['Mineral name'] = k
            new_data.append(new_dic)

        temp_df = pd.DataFrame.from_records(new_data)
        final_df = final_df.append(temp_df)
```

```python
final_df.info(max_cols = 200)
```

```python
final_df=final_df.set_index('Mineral name')
```

```python
final_df
```

```python
final_df.drop_duplicates(inplace=True)
```

```python
final_df.to_csv('../data/webmineral/mineral-5k.csv')
```
